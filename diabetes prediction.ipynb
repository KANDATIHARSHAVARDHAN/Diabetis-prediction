```python
# Converting other values
df['smoking_history'] = df['smoking_history'].replace({'never': 0, 'current': 1})
df
```

```python
# Generating random values of 0 or 1 for 'No Info', 'former', 'ever', 'not current'
categories_to_randomize = ['No Info', 'former', 'ever', 'not current']
random_values = np.random.randint(0, 2, df['smoking_history'].isin(categories_to_randomize).sum())

# Replacing the specified categories with random 0 or 1
df.loc[df['smoking_history'].isin(categories_to_randomize), 'smoking_history'] = random_values
```

```python
y=df['diabetes']
x=df.drop(columns=['diabetes'],axis=1)
x
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
x['gender']=le.fit_transform(x['gender'])
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from tensorflow.keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
```

```python
x.shape
```

```python
model.add(Dense(units=20,activation='relu',kernel_initializer='random_uniform'))
model.add(Dense(units=40,activation='relu',kernel_initializer='random_uniform'))
model.add(Dense(units=45,activation='relu',kernel_initializer='random_uniform'))
model.add(Dense(units=40,activation='relu',kernel_initializer='random_uniform'))
model.add(Dense(units=45,activation='relu',kernel_initializer='random_uniform'))
model.add(Dense(units=1,activation='sigmoid',kernel_initializer='random_uniform'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
x_train.shape
```

```python
model.fit(x_train,y_train,batch_size=32,epochs=40)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred_class = (y_pred > 0.5).astype(int)  # For binary classification
accuracy = accuracy_score(y_pred_class, y_test)
print(accuracy)
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Sample input
sample_input = ['Female', 44.0, 0, 0, 0, 19.31, 6.5, 200]

# Initialize LabelEncoder and StandardScaler
le = LabelEncoder()
sc = StandardScaler()

# Fit the LabelEncoder on the training data (example categories: 'Female' and 'Male')
le.fit(['Female', 'Male'])

# Assume this is your training data for the numerical features
numerical_training_data = np.array([
    [80.0, 0, 1, 0, 25.19, 6.6, 140],
    
    # Add more training data here
])

# Fit the StandardScaler on the training numerical data
sc.fit(numerical_training_data)

# Now, transform the sample input
encoded_categorical = le.transform([sample_input[0]])  # Encoded categorical feature

# Transform the numerical features
numerical_features = np.array(sample_input[1:]).reshape(1, -1)
scaled_numerical = sc.transform(numerical_features)  # Scaled numerical features

# Combine the transformed categorical and scaled numerical features
transformed_input = np.concatenate([encoded_categorical, scaled_numerical.flatten()])

# Convert to a 2D array (as input for a model)
transformed_input = transformed_input.reshape(1, -1)

# Output the transformed array
print(transformed_input)
```
